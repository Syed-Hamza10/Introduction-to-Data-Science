## 1. Read a  Excel File in Pandas Dataframe
>Pandas is a popular Python library used for working in tabular data (similar to the data stored in a spreadsheet). Pandas provides helper functions to read data from various file formats like CSV, Excel spreadsheets, HTML tables, JSON, SQL, and more.

### Reading Excel File

Filename is **big_mart_sales.xlsx** residing in directory name datasets which is in current working directory

After checking the type of data You can see that the data is stored in type DataFrame

```python
# To view the data, we can use the head(n) function. If no parameter is passed it displays top 5 rows

df.head()
```

## 2. Reading a Excel File having Comments in the beginning
You may get an error while reading a Excel file because someone may have added few comments on the top of the file. In pandas we can still read the data set by skipping few rows from the top.

```python
# As a proof of concept Read the file big_mart_sales_comments.xlsx present in the datasets directory.
# Reading this csv file will give you errors as it has few comments in the beginning of file

# read the data without skiprows
sheet_with_comments = pd.read_excel('datasets/big_mart_sales_comments.xlsx')

# view top rows of the data, that is somehow messed up
sheet_with_comments.head()
```

#### Use skiprows parameter to read_csv() function
- To deal with the ParseError, open the Excel file and check if you have some comments on the top.
- If yes, then count the number of rows to skip.
- While reading file, pass the parameter skiprows = n (number of rows to skip)

```python
# we have seen that our csv file contain comments, so we need to skip those rows having comments
# read the dataset using skiprows parameter

# read the data with skiprows parameter
sheet_with_comments = pd.read_excel('datasets/big_mart_sales_comments.xlsx',skiprows=3)

# view the top rows of the data
sheet_with_comments.head()
```

## 3. Reading a portion of Excel File in a Dataframe
- Suppose the dataset inside the Excel file is too big and you dont want to spend that much time for reading that data
- Or might be your system crashes, when you try to load that much data
- solution is read
    - Specific number of rows
    - Specific number of columns

```python
# As a proof of concept let us read just first 100 rows from the csv file
# In the read_excel() function, use the parameter nrows for this purpose


specific_rows = pd.read_excel('datasets/big_mart_sales.xlsx', nrows=50)

# check the shape of the data
specific_rows.shape
```

```python
# If the dataset has large number of columns, and you do not want to load all of them
# In read_excel() function, use the parameter usecols for this purpose
specific_cols = pd.read_excel('datasets/big_mart_sales.xlsx', usecols= ['Item_Identifier', 
                                                                             'Item_Type',
                                                                             'Item_MRP',
                                                                             'Item_Outlet_Sales'])


# check the shape of the data
specific_cols.shape
# view the data
#specific_cols.head()
```

```python
# Ofcourse you can use both the parameters at the same time
fd = pd.read_excel('datasets/big_mart_sales.xlsx', nrows = 25, usecols= ['Item_Identifier', 
                                                                             'Item_Type',
                                                                             'Item_MRP',
                                                                             'Item_Outlet_Sales'])


# check the shape of the data
fd.shape
# view the data
#fd.head()
```

## 4. Reading Excel Workbook with Multiple Sheets
- Suppose we have an Excel file having multiple sheets.
- By default read_excel() function read only the first sheet

```python
# As a proof of concept, we have a excel file big_mart_sales_with_multiple_sheets.xlsx
# This workbook contains different sheets for different years 1985, 1987, 1997... of data
# read the excel sheet with multiple sheets
data_with_multiple_sheets = pd.read_excel('datasets/big_mart_sales_with_multiple_sheets.xlsx')

# if you check/view the data you can see, it only contains the data of first excel sheet (for the year 1985)
data_with_multiple_sheets
```

### a. Let us read all the Sheets one after another by giving their names

```python
sheet_1985 = pd.read_excel('datasets/big_mart_sales_with_multiple_sheets.xlsx',sheet_name='1985')
sheet_1987 = pd.read_excel('datasets/big_mart_sales_with_multiple_sheets.xlsx',sheet_name='1987')
sheet_1997 = pd.read_excel('datasets/big_mart_sales_with_multiple_sheets.xlsx',sheet_name='1997')
```

```python
sheet_1985.shape
```

```python
sheet_1987.shape
```

```python
sheet_1997.shape
```

### b. Use the concat() Function to Make a Single Dataframe

```python
# now you can use the concat method of pandas to concatenate the dataframes

# create an array of dataframes
#sheets = [sheet_1985, sheet_1987, sheet_1997]
# concatenate the array
final_data = pd.concat([sheet_1985, sheet_1987, sheet_1997])

# now you can check the unique sheets in data again
# unique years in final data
final_data.Outlet_Establishment_Year.unique()
```

```python
final_data
```

#### Note the total number of rows in this dataframe equals to 3325 = 1463+932+930 

### c. Store the Contents of this Dataframe to a Single Excel Sheet

```python
# you can store the concatenated data inside your dataframe into a single Excel file
# mention the index= false because we don't want to store the index 1,2,3,... in the Excel file.

final_data.to_excel('temp.xlsx', index=False)

#final_data.to_excel('temp.xlsx', index=False, sheet_name = 'Students')

# you can check datasets directory, the datasets/temp.xlsx file has been created
data = pd.read_excel('temp.xlsx')
#data = pd.read_excel('temp.xlsx', sheet_name = 'Students')

# checking type of data
# You can see that the data is stored in type DataFrame
data
```
