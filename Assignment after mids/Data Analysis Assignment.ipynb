
# Data Loading and Inspection

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
#loads dataset
iris = load_iris()
```

```python
#converts to a DataFrame
data = pd.DataFrame(iris.data, columns = iris.feature_names)
```

# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
# Plot histogram
plt.hist(data['sepal length (cm)'], bins=50, color='skyblue', edgecolor='black')


plt.title('Histogram of Sepal Length', fontdict={'fontsize': 16, 'fontweight': 'bold', 'fontfamily': 'serif'})
plt.xlabel('Sepal Length (cm)', fontdict={'fontsize': 14, 'fontweight': 'normal', 'fontfamily': 'serif'})
plt.ylabel('Frequency', fontdict={'fontsize': 14, 'fontweight': 'normal', 'fontfamily': 'serif'})

# Show plot
plt.show()
```

```python
plt.scatter(data['sepal length (cm)'], data['sepal width (cm)'] , c = data['target'])
plt.legend(labels=data['target'], title='Species')

plt.title('Relationship between Sepal Length and Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Show plot
plt.show()
```

```python
plt.boxplot([data[data['target'] == i]['petal length (cm)'] for i in sorted(data['target'].unique())],
            labels=data['target'].unique())

plt.title('Distribution of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
```

```python
import seaborn as sns

# Selecting only numerical columns
numerical_columns = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

# Create pairplot
sns.pairplot(data, vars=numerical_columns)

# Show plot
plt.show()
```
